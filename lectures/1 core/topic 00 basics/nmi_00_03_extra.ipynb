# 00.03 basics : more examples

##### example 01

show $x^3$ is forward stable.

ie, there exists $\eta > 0$ such that $||\Delta y|| \le \eta\cdot ||y||$ where
<br/>

$$
\Delta y = y - \hat{y} \text{, where}\quad \left\{\quad
\begin{array} &
  \phi : \mathbb{R} \rightarrow \mathbb{R}, x \overset{\phi}{\mapsto} x^3 &\Rightarrow y = \phi(x) \\
  \hat{\phi} : \mathbb{FP} \rightarrow \mathbb{FP}, \hat{x} \overset{\hat{\phi}}{\mapsto} \hat{x}^3 &\Rightarrow \hat{y} = \hat{\phi}(\hat{x}).
\end{array}
\right.
$$

\begin{align}
  \hat y &= \hat x^3 = (\color{blue}{\hat x} \otimes \color{orange}{\hat x}) \otimes \color{green}{\hat x} \\ \\
  &= (\color{blue}{x(1+\delta_x)} \color{red}{\otimes} \color{orange}{x(1+\delta_x)}) \color{purple}{\otimes} (\color{green}{x(1+\delta_x)}) \\ \\
  &= \color{red}{x^2(1+\delta_x)^2(1+\delta_{\otimes})} \color{purple}{\otimes} (\color{green}{x(1+\delta_x)}) \\ \\
  &= \color{purple}{x^3(1+\delta_x)}\color{sandybrown}{^3} \color{red}{(1+\delta_{\otimes})}\color{sandybrown}{^1} \color{purple}{(1+\delta_{\otimes})}\color{sandybrown}{^1} \\ \\
  &= x^3(1+\theta\color{sandybrown}{_5}) \quad\text{by theorem 01} \\
  \\
  &\Downarrow \\
  \\
  \Delta y &= \hat y - y = x^3(1 + \theta\color{sandybrown}{_5}) - x^3 = x^3\theta\color{sandybrown}{_5} \\
  \\
  &\Downarrow \\
  |\Delta y| &= |\hat y - y| \le |x^3||\theta\color{sandybrown}{_5}| \le |y|\cdot\gamma\color{sandybrown}{_5} \quad\text{ by theorem 01, where }\gamma\color{sandybrown}{_5} = \tfrac{n\mu_M}{1-n\mu_M} = \tfrac{\color{sandybrown}{5}\mu_M}{1-\color{sandybrown}{5}\mu_M} \\
  \\
  &\text{choose } \eta = \gamma\color{sandybrown}{_5} \text{, then } |\hat y - y| \le \eta\cdot |y|. \enspace\blacksquare
\end{align}

##### example 02

show $x^3$ is numerically stable.

ie, there exists $\eta > 0$ and $\epsilon > 0$ such that $||\Delta y|| \le \eta\cdot ||y||$ and $|\Delta x| \le \epsilon |x|$ where
<br/>

$$
\Delta y = y - \hat{y} \text{, where}\quad \left\{\quad
\begin{array} &
  \phi : \mathbb{R} \rightarrow \mathbb{R}, x \overset{\phi}{\mapsto} x^3 &\Rightarrow y = \phi(x) \\
  \hat{\phi} : \mathbb{FP} \rightarrow \mathbb{FP}, \hat{x} \overset{\hat{\phi}}{\mapsto} \hat{x}^3 &\Rightarrow \hat{y} = \hat{\phi}(\hat{x}).
\end{array}
\right.
$$

note: $\enspace \hat y = x^3(1 + \theta_5), \gamma_5$ from example 01, forward stability.

\begin{align}
  \hat y + \Delta y &= \phi(x+\Delta x) \quad\text{see highams schematic} \\
  \\
  x^3(1 + \theta_5) + \Delta y &= (x+\Delta x)^3 = x^3 + 3x^2(\Delta x) + 3x(\Delta x)^2 + (\Delta x)^3 \\
  \\
  \Rightarrow \Delta y &= 3x^2 (\Delta x) + 3x(\Delta x)^2 + (\Delta x)^3 - x^3\theta_5. \\
  \\
  \text{what } \boldsymbol{\color{blue}{\eta}},\boldsymbol{\color{darkorange}{\epsilon}} &\text{ such that } |\Delta y| \le \eta\cdot |y|,|\Delta x| \le \epsilon\cdot |x| \text{ ?} \\
  \\
  |\Delta y| &\le 3|x|^2|\Delta x| + 3|x||\Delta x|^2 + |\Delta x|^3 + |x|^3\theta_5 \\
  \\
  &\le 3|x|^2\epsilon|x| + 3|x|\epsilon^2|x|^2 + \epsilon^3|x|^3 + |x|^3\gamma_5 \quad\text{ by definition of }\epsilon, \gamma_5 \\
  \\
  &\le 3\epsilon|x|^3 + 3\epsilon^2|x|^3 + \epsilon^3|x|^3 + |x|^3\gamma_5 = (3\epsilon + 3\epsilon^2 + \epsilon^3 + \gamma_5)|y| \\
  \\
  \Rightarrow |\Delta y| &\le \eta\cdot |y| \Rightarrow \eta = 3\epsilon + 3\epsilon^2 + \epsilon^3 + \gamma_5. \enspace\checkmark
\end{align}

ie, take any $\boldsymbol{\color{darkorange}{\epsilon > 0}}$ and any $\Delta x$ such that $|\Delta x| \le \epsilon|x|$. then $\boldsymbol{\color{blue}{\eta = 3\epsilon + 3\epsilon^2 + \epsilon^3 + \gamma_5}}$ and $\Delta y = 3x^2(\Delta x) + 3x(\Delta x)^2 + (\Delta x)^3 - x^3\theta_5$ and by construction, $|\Delta y| \le \eta\cdot |y|$ and $\hat{y} + \Delta y = \phi(x+\Delta x)$. ie, $\hat{\phi}(x) = \hat{x}^3$ is numerically stable. $\blacksquare$

note: thats "any" $\epsilon$ vs "best, smallest" $\epsilon, \eta$. also, the only part of the derived $\eta$ outside control is $\gamma_5$, so choose $\epsilon \ll 1$ such that $\eta$ is barely above $\gamma_5$. ie, $\eta = \gamma_5\cdot c_0$ where $c_0 \ge 1$.



##### example 03

calculate error for $f_1(a,b) = (a+b)^2$.

\begin{align}
  \color{blue}{\hat{a}} \oplus \color{darkorange}{\hat{b}} &= \color{blue}{a(1+\delta_a)} \color{green}{\oplus} \color{darkorange}{b(1+\delta_b)} \\
  \\
  &= (\color{blue}{a} + \color{darkorange}{b} + \color{blue}{a\delta_a} + \color{darkorange}{b\delta_b}) \color{green}{(1+\delta_{\oplus})} \\
  \\
  &= (\color{blue}{a} + \color{darkorange}{b}) \left(1+\frac{\color{blue}{a\delta_a} + \color{darkorange}{b\delta_b}}{\color{blue}{a} + \color{darkorange}{b}}\right) \color{green}{(1+\delta_{\oplus})} \\
  \\
  &= (a+b) \color{red}{(1+\delta_{a,b})} \color{green}{(1+\delta_{\oplus})} \\
  \\
  &\Downarrow \\
  \\
  \hat{y} &= (\hat{a} \oplus \hat{b}) \color{purple}{\otimes} (\hat{a} \oplus \hat{b}) \\
  \\
  &= [(a+b) \color{red}{(1+\delta_{a,b})} \color{green}{(1+\delta_{\oplus})}] [(a+b) \color{red}{(1+\delta_{a,b})} \color{green}{(1+\delta_{\oplus})}]  \color{purple}{(1+\delta_{\otimes})} \\
  \\
  &= (a+b)^2 \color{red}{(1+\delta_{a,b})}^{\color{sandybrown}{2}} \color{green}{(1+\delta_{\oplus})}^{\color{sandybrown}{2}} \color{purple}{(1+\delta_{\otimes})}^{\color{sandybrown}{1}} \\
  \\
  &= (a+b)^2 (1+\theta_{\color{sandybrown}{5}}) \quad\text{theorem 01} \\
  \\
  &\Downarrow \\
  \\
  |\Delta y| &= |y - \hat{y}| = |(a+b)^2 - [(a+b)^2 (1 + \theta_{\color{sandybrown}{5}})] \\
  \\
  &= |(a+b)^2 \theta_{\color{sandybrown}{5}}| = |y|\cdot \theta_{\color{sandybrown}{5}} \le \gamma_{\color{sandybrown}{5}}\cdot |y|.
\end{align}
