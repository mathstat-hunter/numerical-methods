## accumulated rounding error examples

wrt stability, typed out with latex.

### ex 01
forward stability. ie, $\eta > 0$ such that $||\Delta y|| \le \eta||y||$.

$\Delta y = \hat y - y$

$\phi: \mathbb{R}\rightarrow\mathbb{R}$, $x \overset{\phi}{\mapsto} x^3$

$\hat\phi: \mathbb{F}\rightarrow\mathbb{F}$, $\hat x \overset{\hat\phi}{\mapsto} \hat x^3 \Rightarrow \hat y = \hat\phi(\hat x) = \hat x^3$

</br>
\begin{align}
  \hat y &= \hat x^3 = (\color{red}{\hat x} \otimes \color{blue}{\hat x}) \otimes \color{green}{\hat x} \\ \\
  &= (\color{red}{x(1+\delta_x)} \color{orange}{\otimes} \color{blue}{x(1+\delta_x)}) \color{purple}{\otimes} (\color{green}{x(1+\delta_x)}) \\ \\
  &= \color{orange}{x^2(1+\delta_x)^2(1+\delta_{\otimes})} \color{purple}{\otimes} (\color{green}{x(1+\delta_x)}) \\ \\
  &= \color{purple}{x^3(1+\delta_x)}\color{teal}{^3} \color{orange}{(1+\delta_{\otimes})}\color{teal}{^1} \color{purple}{(1+\delta_{\otimes})}\color{teal}{^1} \\ \\
  &= x^3(1+\theta\color{teal}{_5}) \textrm{ [theorem 01]}
\end{align}

</br>
$\hat y - y = x^3(1 + \theta\color{teal}{_5}) - x^3 = x^3\theta\color{teal}{_5}$

$|\hat y - y| = |x^3||\theta\color{teal}{_5}| \le |y|\gamma\color{teal}{_5} \textrm{ [theorem 01]}$

$\gamma\color{teal}{_5} = \frac{\color{teal}{n}\mu_M}{1-\color{teal}{n}\mu_M} = \frac{\color{teal}{5}\mu_M}{1-\color{teal}{5}\mu_M}$.

let $\eta = \gamma\color{teal}{_5}$, then $|\hat y - y| \le \eta|y|$. $\checkmark$



</br></br></br></br></br>


### ex 02
numerical stability. ie, $\eta > 0$ such that $|\Delta y| \le \eta |y|$ and $\epsilon > 0$ such that $|\Delta x| \le \epsilon |x|$.

$y = \phi(x) = x^3$. need $\Delta x$, $\Delta y$.

note: $\hat y = x^3(1 + \theta_5), \gamma_5$ from eg 01.

</br>
\begin{align}
  \hat y + \Delta y &= \phi(x+\Delta x) \\
  x^3(1 + \theta_5) + \Delta y &= (x+\Delta x)^3 = x^3 + 3x^2(\Delta x) + 3x(\Delta x)^2 + (\Delta x)^3 \\
  \Rightarrow \Delta y &= 3x^2 (\Delta x) + 3x(\Delta x)^2 + (\Delta x)^3 - x^3\theta_5
\end{align}


</br>

what <font color=red>$\eta$</font> for $|\Delta y| \le \eta|y|$?

$\Rightarrow$ what <font color=blue>$\epsilon$</font> for $|\Delta x| \le \epsilon|x|$?

</br>
\begin{align}
  |\Delta y| &\le 3|x|^2|\Delta x| + 3|x||\Delta x|^2 + |\Delta x|^3 + |x|^3\theta_5 \\
  &\le 3|x|^2\epsilon|x| + 3|x|\epsilon^2|x|^2 + \epsilon^3|x|^3 + |x|^3\gamma_5 \\
  &\Rightarrow 3\epsilon|x|^3 + 3\epsilon^2|x|^3 + \epsilon^3|x|^3 + |x|^3\gamma_5 \Rightarrow (3\epsilon + 3\epsilon^2 + \epsilon^3 + \gamma_5)|y| \\
  \\
  \Rightarrow |\Delta y| &\le \eta|y| \Rightarrow \eta = 3\epsilon +3\epsilon^2 + \epsilon^3 + \gamma_5
\end{align}

</br>

ie, <font color=blue>take any</font> $\color{blue}{\epsilon > 0}$, any $\Delta x$ such that $|\Delta x| \le \epsilon|x|$. then let <font color=red>$\eta = 3\epsilon + 3\epsilon^2 + \epsilon^3 + \gamma_5$</font>, $\Delta y = 3x^2(\Delta x) + 3x(\Delta x)^2 + (\Delta x)^3 - x^3\theta_5$.

by construction (ie, by choice), we know $|\Delta y| \le \eta|y|, \hat y+\Delta y = \phi(x+\Delta x)$. $\checkmark$

ie, $\hat\phi(x) = \hat x^3$ is numerically stable.

also, "any" $\epsilon$ vs "the best, smallest" $\epsilon$, $\eta$. note that the only part of the derived $\eta$ that you cannot control is $\gamma_5$, so choose $\epsilon \ll 1$ and $\eta$ will barely above $\gamma_5$. ie, $\eta = \gamma_5\cdot c_0$ where $c_0 \ge 1$.

### ex 03

error term for $y = f_1(a,b) = (a+b)^2$.


\begin{align}
  a\oplus b &= a(1+\delta_a)\oplus b(1+\delta_b) \\
  &= (a+b+a\delta_a + b\delta_b)(1+\delta_{\oplus}) \\
  &= (a+b)\left(1+\frac{a\delta_a + b\delta_b}{a+b}\right)(1+\delta_{\oplus}) \\
  &= (a+b)(1+\delta_{a,b})(1+\delta_{\oplus}) \\
  \\
  &\Downarrow \\
  \\
  \hat{y} &= (a\oplus b)\otimes(a\oplus b) \\
  &= [(a+b)(1+\delta_{a,b})(1+\delta_{\oplus})][(a+b)(1+\delta_{a,b})(1+\delta_{\oplus})](1+\delta_{\otimes}) \\
  &= (a+b)(1+\delta_{a,b})^2(1+\delta_{\oplus})^2(1+\delta_{\otimes}) \\
  &= (a+b)(1+\theta_5)
\end{align}
