      {[(x_2-x_1)^2+(y_2-y_1)^2]^{\frac{3}{2}}} +
    \frac{g\,m_3(x_3-x_1)}
      {[(x_3-x_1)^2+(y_3-y_1)^2]^{\frac{3}{2}}} \\
