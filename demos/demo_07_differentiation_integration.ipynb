### q3. (10 pts) use romberg integration to approximate the definite integral. compare with the correct integral and report error.

<b><font color=orange>CHOOSE ONE</font></b>
</br>


#### (a) apply approximation $R_{33}$ <b><font color=green>by hand only</font></b> to integral

$$\int_0^{\frac{\pi}{2}} cos \text{ }x \text{ } dx.$$
</br>


#### (b) apply approximation $R_{55}$ with <b><font color=red>code</font></b> to integral

$$\int_0^1 \frac{x^3 \text{ }dx}{x^2+1}.$$
</br>


##### example


$\int_0^4 \frac{x \text{ }dx}{\sqrt{x^2+9}}$


```python
import numpy as np
import scipy as sp

f = lambda x: x/np.sqrt(pow(x,2)+9)
a = 0
b = 4

def main():
  d = 16 # pow(2,k)
  h = (b-a)/d
  x = np.linspace(a,b,d+1) # 2^k + 1 points ~ k = 4 ~ r[5,5]
  y = f(x)
  rc = sp.integrate.romb(y,dx=h,show=True)

if __name__ == "__main__":
  main()
```
