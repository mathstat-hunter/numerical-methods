```python
# nmi 2024 : hw10 : q2, demonstration of steps ~ NOT THE ANSWER ASKED!!

import numpy as np
from tabulate import tabulate

# solvers
step_euler = lambda h,f,t,w: w + h*f(t,w)
step_trapezoid = lambda h,f,t,w: w + (h/2)*(f(t,w) + f(t+h, w + h*f(t,w)))
step_midpoint = lambda h,f,t,w: w + h*f(t+h/2, w + h/2*f(t,w))
def step_rk4(h,f,t,w):
  s1 = f(t,w)
  s2 = f(t+h/2, w + h/2*s1)
  s3 = f(t+h/2, w + h/2*s2)
  s4 = f(t+h  , w + h*s3)
  w = w + h/6*(s1 + 2*s2 + 2*s3 + s4)
  return w

# problem
def main():

  if True: # hack to hide question choice options
    # y, actual # for error checking
    y_q1a = lambda t: np.exp(pow(t,3)/3)
    y_q1b = lambda t: np.exp(pow(t,2)+2*t)
    y_q1c = lambda t: np.exp(pow(t,5))
    y_q1d = lambda t: pow(3*t+1,1/3)
    y_act = [y_q1a,y_q1b,y_q1c,y_q1d]
    # ivp, y' = f(t,y)
    f_q1a = lambda t,y: (2*t+pow(t,4))*y
    f_q1b = lambda t,y: (4*pow(t,2) + 8*t + 6)*y
    f_q1c = lambda t,y: (20*pow(t,3) + 25*pow(t,8))*y
    f_q1d = lambda t,y: 10/pow(y,8)
    ydot = [f_q1a,f_q1b,f_q1c,f_q1d]
    # ivp, initial condition
    ic_q1a = np.array([1,0])
    ic_q1b = np.array([1,2])
    ic_q1c = np.array([1,0])
    ic_q1d = np.array([1,1,-2])
    ic = [ic_q1a,ic_q1b,ic_q1c,ic_q1d]
    # identifier, ivp
    q1a=0;q1b=1;q1c=2;q1d=3

    # method
    step = [step_euler,step_trapezoid,step_midpoint,step_rk4]
    # identifier, method
    euler=0;trapezoid=1;midpoint=2;rk4=3
  q = q1a
  method = euler

  # ivp
  def f(t,y): # y' ≡ f(t,y), model, system of equations
    dy = np.zeros_like(y)
    for i in range(len(y)-1):
      dy[i] = y[i+1]
    dy[-1] = ydot[q](t,y[0])
    return dy
  w0 = ic[q] # initial value # not via y_act bc not part of ivp
  tt = [0,1] # interval, same for all

  # grid
  h = 0.0001 # STEPSIZE
  n = int((tt[1]-tt[0])/h) # points, excluding 0

  # tabulate, compensate for lack of skip wrt rows displayed
  n_print = 10
  n_skip = int(n/n_print)

  # data, initialize
  t0 = tt[0]; y0 = y_act[q](t0)
  data = np.zeros((n_print+1,5)) # i,t,y,w,e
  data[0,:] = [0,t0,y0,w0[0],abs(w0[0]-y0)]

  # calculate
  for i in range(1,n+1):
    ti = h*i
    wi = step[method](h,f,t0,w0)
    yi = y_act[q](ti)

    if i % n_skip == 0: # compensate for lack of skip wrt rows displayed
      data[int(i/n_skip):] = [i,ti,yi,wi[0],abs(yi-wi[0])]
    t0 = ti
    w0 = wi

  print()
  print(tabulate(data,headers=["i","t","y","w","e"],floatfmt=(".0f",".8f",".8f",".8f",".8f"),tablefmt="github"))

if __name__ == "__main__":
  main()
```

### q3 (10 pts) classical physics.


express, as the solution of a second-order boundary value problem, the height of a projectile that is thrown from the top of a $60$-meter tall building and takes $5$ seconds to reach the ground. solve the problem (this can be done without code) and find the maximum height reached by the projectile.

```python
import scipy as sp
import textwrap

def main():
  tf = 5; # time, final
  h0 = 60 # height, initial
  v0 = (pow(tf,2)/2*sp.constants.g - h0)/tf # velocity, initial: solve y(tf) = 0
  t_hmax = v0/sp.constants.g # solve y'(t_hmax) = 0 = -g*t_hmax + v0
  y_hmax = -sp.constants.g/2*pow(t_hmax,2) + v0*t_hmax + h0 # evaluate y(t_hmax)

  s_blahblahblah = f"for projectile to have {tf} seconds air time against the \
    acceleration of gravity, it must have initial vertical velocity of \
    {v0:.4f} m/s and its trajectory was such that it peaked at {y_hmax:.4f} \
    meters at {t_hmax:.4f} seconds."
  print(textwrap.fill(" ".join(s_blahblahblah.split()),width=70))

if __name__ == "__main__":
  main()
```

#### a) (10 pts) provide discretized form of linear BVP.

\begin{align}
  \left\{
    \begin{array}
      && y'' = y + \frac{2}{3}e^t \\
      & y(0) = 0 \\
      & y(1) = \frac{1}{3}e.
    \end{array}
  \right.
\end{align}
</br>


\begin{align}
  y'' &= y + \frac{2}{3}e^t \\
  \\
  &\Downarrow \\
  \\
  \text{centered diff, 2nd deriv}\qquad y'' &\approx \frac{y(t+h) - 2y(t) + y(t-h)}{h^2} \\
  &= \frac{y_{i+1} - 2y_i + y_{i-1}}{h^2} \\
  &\Rightarrow \frac{w_{i+1} - 2w_i + w_{i-1}}{h^2} \\
  \\
  &\Downarrow \\
  \\
  \frac{y_{i+1} - 2y_i + y_{i-1}}{h^2} &= w_i + \frac{2}{3}e^t \\
  \\
  &\Downarrow \\
  \\
  \frac{w_{i+1} - 2w_i + w_{i-1}}{h^2} - (w_i + \frac{2}{3}e^t) &= 0 \\
  \\
  &\Downarrow \quad\text{answer} \\
  \\
  \text{general}\quad &w_{i+1} - (2+h^2)w_i + w_{i-1} - h^2\frac{2}{3}e^t = 0 \\
  \\
  \text{at }t=a,\quad &y_a - (2+h^2)w_1 + w_2 - h^2\frac{2}{3}e^a = 0, \qquad a = 0, y_a = 0 \\
  \\
  \text{at }t=b,\quad &w_{n-1} - (2+h^2)w_n + y_b - h^2\frac{2}{3}e^b = 0, \qquad b = 1, y_b = \frac{1}{3}e \\
