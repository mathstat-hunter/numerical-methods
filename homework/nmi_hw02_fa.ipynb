### global functions, variables


````python
# debug
gb_debug = True

# limits, programmatic
gepsilon = 0.00001 # max error
gimax = 100 # max iterations

# x interval
gx_from = 0.
gx_thru = 1.

# x, educated guess
gx0 = 0.60

def f(ax):
  rc = ax**3 + ax - 1.
  if gb_debug:
    print(f"inside f: f({ax}) = {rc}")
  return rc
````

#### code

````python
def f_bisection(aa,ab,aepsilon):
  # biiig assumption that f(a)*f(b) < 0
  a = aa;b = ab # habit to save args if changing them
  fa = f(a)
  isteps = 0

  while abs((b-a)/2.) > aepsilon:
    c = (a+b)/2.
    fc = f(c)
    isteps += 1 # common shortcut for isteps = isteps + 1
    if fc == 0:
      break
    elif fa*fc < 0:
      b = c
    else:
      a = c
  return c,isteps
````

````python
root,steps = f_bisection(gx_from,gx_thru,gepsilon)
print(f"given max error {gepsilon:.5f}, found root = {root:.5f} at {steps} iterations")
````

### fixed-point

<font color = "blue">not needed for homework; for continuity of methods.</font>

#### algorithm

$x_{i+1} = x_{i} - f(x_{i})$


#### code, 230905 static f(x) and <font color=red>what can go wrong</font>

````python
# fixed point
def f_fpi(ax0,aepsilon,aimax):
  rc = 0
  isteps = 0
  x = ax0
  fx = f(x)
  while (abs(fx - x) > aepsilon) and (isteps <= aimax):
    x = fx
    fx = f(x)
    isteps += 1 # common shortcut for isteps = isteps + 1
  rc = x
  return rc,isteps
````

##### $g_3(x) = \frac{1+2x^3}{1+3x^2}$

### newton-raphson

#### algorithm

newton-raphson: $x_{new} = x_{old} - \frac{f(x_{old})}{f'(x_{old})}$


```
icount = 0
fx = f(x_old)
while (abs(fx) > epsilon) and (icount <= imax):
  x_new = x_old - f(x_old)/f'(x_old)
  fx = f(x_new)
  x_old = x_new
  icount = icount + 1
return x_old
```



#### code

````python
def df(ax):
  rc = 3.*ax**2 + 1. # f'(x) = d(x^3 + x - 1)
  if gb_debug:
    print(f"inside df({ax}) = {rc}")
  return rc

def f_newton(af,adf,ax,aepsilon,aimax):
  root = 0;iterations = 0

  icount = 0
  x_old = ax
  fx = f(x_old)
  while (abs(fx) > aepsilon) and (icount < aimax): ##
    #x_new = x_old - f(x_old)/f'(x_old)
    x_new = x_old - fx/adf(x_old) # at this point, fx = f(x_old) ~ fx_old
    #fx = f(x_new)
    fx = f(x_new)
    #x_old = x_new
    x_old = x_new
    #icount = icount + 1
    icount += 1

  root = x_old;iterations = icount ##
  return root,iterations
````

### secant method (no derivatives required)

#### algorithm

secant: $x_{i+1} = x_i - f(x_i)\cdot\frac{x_i - x_{i-1}}{f(x_i) - f(x_{i-1})}$
</br></br>




#### code

````python
def qf(af,ax1,ax2):
  # difference quotient here, plz
  rc = (af(ax2)-af(ax1))/(ax2 - ax1)
  if gb_debug:
    print(f"inside qf({ax1},{ax2}) = {rc}")
  return rc

def f_secant(af,ax1,ax2,aepsilon,aimax):
  root = 0;iterations = 0

  # do some stuffs here
  x_old = ax1
  x_older = ax2
  fx_old = af(x_old)
  if abs(fx_old) < aepsilon:
    return x_old,iterations
  fx_older = af(x_older)
  if abs(fx_older) < aepsilon:
    return x_older,iterations

  # dq = (x_old - x_older)/(fx_old - fx_older)
  dq = qf(af,x_old,x_older)
  x_new = x_old - fx_old/dq
  fx = af(x_new)
  iterations = iterations + 1

  while (abs(fx) > aepsilon) and (iterations <= aimax):
    x_older = x_old
    fx_older = fx_old
    x_old = x_new
    fx_old = fx
    # dq = (x_old - x_older)/(fx_old - fx_older)
    dq = qf(af,x_old,x_older)
    x_new = x_old - fx_old/dq
    fx = af(x_new)
    iterations = iterations + 1
  root = x_new

  # dont be like me and forget to put values into root,iteration here
  return root,iterations
````
